打印问卷题目（在初步解析后，可以使用正则表达式进一步处理）

````python
questions = r.html.xpath("//div[@class='field-label']", first=True).text

print(questions)
````

````python
import re

print(re.findall(r'.*\*\n', questions))
````

获取输入框的位置（可以进一步筛选出需要的input位置）

````python
inputs = r.html.find("input", first=False)

print(inputs)
````

````python
for inp in inputs:
  print(inp.html)
````
