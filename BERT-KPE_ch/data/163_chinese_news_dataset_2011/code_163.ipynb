```python
import json

raw_data = []
with open(raw_data_path) as f:
    for data in f:
      raw_data.append(json.loads(data.rstrip(';\n')))
print(raw_data[0])
```

```python
import jieba
# 共13703条数据
all_data = []
with open('/content/drive/MyDrive/paipai-NLP/BERT-KPE_ch/data/163_chinese_news_dataset_2011/preprocessed_163.json', 'w', encoding='utf-8') as ff:
  for data in raw_data:
    doc_id = int(data['id'])
    text_tokens = jieba.lcut(data['summary'][5:])
    keyphrases = data['tags']
    preprocessed_data = {"doc_id":doc_id, "text_tokens":text_tokens, 'keyphrases':keyphrases}
    all_data.append(preprocessed_data)
  json.dump(all_data,ff,ensure_ascii=False)
```

```python
# train 取前13000条
train_data = []
with open('/content/drive/MyDrive/paipai-NLP/BERT-KPE_ch/data/163_chinese_news_dataset_2011/preprocessed_163_train.json', 'w', encoding='utf-8') as f_train:
  for data in raw_data[:13000]:
    doc_id = int(data['id'])
    text_tokens = jieba.lcut(data['summary'][5:])
    keyphrases = data['tags']
    preprocessed_data = {"doc_id":doc_id, "text_tokens":text_tokens, 'keyphrases':keyphrases}
    train_data.append(preprocessed_data)
  json.dump(train_data,f_train,ensure_ascii=False)

# test 取后703条
test_data = []
with open('/content/drive/MyDrive/paipai-NLP/BERT-KPE_ch/data/163_chinese_news_dataset_2011/preprocessed_163_test.json', 'w', encoding='utf-8') as f_test:
  for data in raw_data[13000:]:
    doc_id = int(data['id'])
    text_tokens = jieba.lcut(data['summary'][5:])
    keyphrases = data['tags']
    preprocessed_data = {"doc_id":doc_id, "text_tokens":text_tokens, 'keyphrases':keyphrases}
    test_data.append(preprocessed_data)
  json.dump(test_data,f_test,ensure_ascii=False)
```
